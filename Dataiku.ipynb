## 1 Read all file

```python
import pandas as pd

Oil_link=r'D:\DS_book_linux\13.DWH-Cloud\Daitaku\Core Designer cert\CO2_and_Oil.csv'
Meat_link=r'D:\DS_book_linux\13.DWH-Cloud\Daitaku\Core Designer cert\Meat_and_Egg_Production.csv'
Urban_link=r'D:\DS_book_linux\13.DWH-Cloud\Daitaku\Core Designer cert\Urbanization_GDP_and_Population.csv'
```

```python
oil_data=pd.read_csv(Oil_link)
print(oil_data.shape)
oil_data.head()
```

```python
meat_data=pd.read_csv(Meat_link)
print(meat_data.shape)
meat_data.head()
```

```python
urban_data=pd.read_csv(Urban_link)
print(urban_data.shape)
urban_data.head()
```

```python
# Left join
oil_meat= pd.merge(oil_data,meat_data,how='left',
left_on=['Entity','Code','Year'], right_on = ['Entity','Code','Year'])
print(oil_meat.shape)
oil_meat_urban= pd.merge(oil_meat,urban_data,how='left',
left_on=['Entity','Code','Year'], right_on = ['Entity','Code','Year'])
print(oil_meat_urban.shape)
oil_meat_urban.head()
```

## 2 Please analyze only data from 2008-2012 (inclusive).

```python
oil_meat_urban_filter= oil_meat_urban[(oil_meat_urban['Year']<=2012) & (oil_meat_urban['Year'] >=2008)]
print(oil_meat_urban_filter.shape)
oil_meat_urban_filter.head()
```

```python
oil_meat_urban_filter.columns
```

## 3 Some columns report country total data. Transform these columns into per capita data. You will not need the original columns (including Population) for further analysis.

```python
oil_meat_urban_filter.loc[:,'Oil_per_Cap']=oil_meat_urban_filter.loc[:,'Oil production (Etemad & Luciana) (terawatt-hours)']/oil_meat_urban_filter.loc[:,'Population']
oil_meat_urban_filter.loc[:,'meat_prod_tonnes_Cap']=oil_meat_urban_filter.loc[:,'meat_prod_tonnes']/oil_meat_urban_filter.loc[:,'Population']
oil_meat_urban_filter.loc[:,'Food Balance Sheets_Cap']=oil_meat_urban_filter.loc[:,'Food Balance Sheets: Eggs - Production (FAO (2017)) (tonnes)']/oil_meat_urban_filter.loc[:,'Population']
oil_meat_urban_filter.loc[:,'meat_prod_tonnes_Cap']=oil_meat_urban_filter.loc[:,'meat_prod_tonnes']/oil_meat_urban_filter.loc[:,'Population']
oil_meat_urban_filter.loc[:,'Urban (%)_Cap']=oil_meat_urban_filter.loc[:,'Urban (%)']/oil_meat_urban_filter.loc[:,'Population']
```

## 4 Several columns of the per capita data have missing rows. For now, fill these values with 0’s.

```python
oil_meat_urban_filter.info()
```

```python
oil_meat_urban_filter_fillNA=oil_meat_urban_filter.fillna(0)
```

```python
oil_meat_urban_filter_fillNA_selected=oil_meat_urban_filter_fillNA[['Entity','Year','Per capita CO₂ emissions (tonnes per capita)','Meat_supply_kgperCap',
'GDP per capita (int.-$) ($)','Oil_per_Cap','meat_prod_tonnes_Cap','Food Balance Sheets_Cap','Urban (%)_Cap','Urban (%)']]
```

```python
print(oil_meat_urban_filter_fillNA_selected.shape)
oil_meat_urban_filter_fillNA_selected.head()
```
